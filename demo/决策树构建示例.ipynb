# 决策树建模例子

简单创建一个数据集,两个特征,一个目标变量,共两行数据

```python
X = [[0, 0], [1, 1]]
Y = [0, 1] 
```

使用sklearn中的tree模块,设定max_depth=3(最大树深度为3)

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier(max_depth=3)
clf = clf.fit(X, Y)
```

预测一个新数据的类别,预测类别为1

```python
clf.predict([[2., 2.]]) 
```

预测一个新数据的类别的概率(类0为0,类1为100%)

```python
clf.predict_proba([[2., 2.]]) 
```

使用treeplot画出决策树结构

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(10,5))
tree.plot_tree(clf) 
```

上述树中
+ 当X[0]小于等于0.5时,预测结果为类别0(value中,第一个元素为类别0的频数)
+ 当X[0]大于0.5时,预测结果为类别1(value中,第二个元素为类别1的频数)
